## Question 1:
Write a program that takes a string as input, and counts the frequency of each word in the string, there might be repeated characters in the string. Your task is to find the highest frequency and returns the length of the highest-frequency word. 

Note - You have to write at least 2 additional test cases in which your program will run successfully and provide an explanation for the same. Example input - string = “write write write all the number from from from 1 to 100” Example output - 5 Explanation - From the given string we can note that the most frequent words are “write” and “from” and the maximum value of both the values is “write” and its corresponding length is 5

```python
input_string = input()
input_string = input_string.split()
```

```python
print(input_string)
```

```python
frequency_dict = {}
for item in input_string:
    frequency_dict[item] = frequency_dict.get(item, 0) + 1
```

```python

def find_highest_key_value(dictionary):
    highest_value = max(dictionary.values())
    highest_pairs = []

    for key, value in dictionary.items():
        if value == highest_value:
            highest_pairs.append((key, value))

    return highest_pairs


key_value_pairs = find_highest_key_value(frequency_dict)

values = [value for key, value in key_value_pairs]
keys = [key for key, value in key_value_pairs]
print(f'From the given string we can note that the most frequent words are {keys} and its corresponding length is {values}')

```

## Question 2: -

Consider a string to be valid if all characters of the string appear the same number of times. It is also valid if
he can remove just one character at the index in the string, and the remaining characters will occur the same
number of times. Given a string, determine if it is valid. If so, return YES , otherwise return NO .
Note - You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.
Example input 1 - s = “abc”. This is a valid string because frequencies are { “a”: 1, “b”: 1, “c”: 1 }
Example output 1- YES
Example input 2 - s “abcc”. This string is not valid as we can remove only 1 occurrence of “c”. That leaves
character frequencies of { “a”: 1, “b”: 1 , “c”: 2 }
Example output 2 - NO

```python
def isvalidstring(s):
    frequency = {}
    for char in s:
        frequency[char] = frequency.get(char, 0) + 1

    common_frequency = max(frequency.values())

    count_diff_freq = sum(1 for val in frequency.values() if val != common_frequency)

    if count_diff_freq == 0 or (count_diff_freq == 1 and (common_frequency - 1) in frequency.values()):
        return "YES"
    else:
        return "NO"
```

```python
print(isvalidstring("abc"))    
```

```python
print(isvalidstring("abcc"))   
```

```python
def merge_dicts(dict1, dict2):
    if pd.isna(dict1):
        return dict2
    elif pd.isna(dict2):
        return dict1
    elif isinstance(dict1, dict) and isinstance(dict2, dict):
        return {**dict1, **dict2}
    else:
        return {}

```

### 4.  Get the episode name from each season with the highest and lowest rating

```python
grouped_seasons = data_tvmaze.groupby('season')

# Iterating over each season group
for season, season_group in grouped_seasons:

    # Find the episode with the highest rating in the season
    highest_rating_episode = season_group.loc[season_group['rating'].idxmax(), 'name']
    
    # Find the episode with the lowest rating in the season
    lowest_rating_episode = season_group.loc[season_group['rating'].idxmin(), 'name']
    
    print(f"Season {season}: Highest Rating Episode - {highest_rating_episode}, Lowest Rating Episode - {lowest_rating_episode}")
```
